# Earned rewards per agent type


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import os, sys, glob, ast
```

```python
os.chdir('C:\\Users\\annae\\Desktop\\pymdp_depression\\rewards_simulation_results\\revision')
df_c = pd.read_csv('earned_rewards_coop_rev.csv')
df_h = pd.read_csv('earned_rewards_host_rev.csv')

df_c = df_c.loc[:, ~df_c.columns.str.contains('^Unnamed')]
df_h = df_h.loc[:, ~df_h.columns.str.contains('^Unnamed')]

xlabels = ['Healthy', 'A: uncertainty', 'B: fatalistic', 'C: loss averse', 'D: pessimistic', 
          'Depression: insens.-pess.', 'Depression: fatalistic', 'SAD: insecure-avoidant', 'SAD: trauma-defeat']

xlabels_rev = ['Healthy','Depression: insens.-pess.', 'Depression: fatalistic', 'SAD: insecure-avoidant', 'SAD: trauma-defeat']

df_c = df_c.rename(columns=dict(zip(df_c.columns, xlabels)))
df_h = df_h.rename(columns=dict(zip(df_h.columns, xlabels)))
```

```python
results_c, results_h = dict(), dict()
names = list(df_c.columns)
std_err_c, std_err_h = [],[]

for player in names: 
    mean = df_c[player].mean()
    results_c[player] = mean
    std_err_c.append(df_c[player].std())
    
for player in names: 
    mean = df_h[player].mean()
    results_h[player] = mean
    std_err_h.append(df_h[player].std())
    
# filter out the transdiagnostic processes that we do not want in the revision. 
results_c_rev = filtered_data = {k: v for k, v in results_c.items() if not (k[0].isupper() and k[1] == ':')}
std_c_rev = std_err_c[:1] + std_err_c[5:]
results_h_rev = filtered_data = {k: v for k, v in results_h.items() if not (k[0].isupper() and k[1] == ':')}
std_h_rev = std_err_h[:1] + std_err_h[5:]
```

```python
fig, axs = plt.subplots(1,2)
fig.suptitle('Average rewards (N=20, T=40)')
axs[0].bar(results_c_rev.keys(), results_c_rev.values(), yerr=std_c_rev, color='lightskyblue')
axs[0].set_xticklabels(xlabels_rev, rotation=90)
axs[0].set_ylabel('Average reward')
axs[0].set_title('cooperative start')

axs[1].bar(results_h_rev.keys(), results_h_rev.values(), yerr=std_c_rev, color='salmon')
axs[1].set_xticklabels(xlabels_rev, rotation=90)
axs[1].set_title('hostile start')

plt.tight_layout()
plt.savefig('earned_rewards_agents.svg')
```

```python
# ALL AGENTS - including transdiagnostic processes. 
fig, axs = plt.subplots(1,2)
fig.suptitle('Average rewards (N=20, T=40)')
axs[0].bar(results_c.keys(), results_c.values(), yerr=std_err_c, color='lightskyblue')
axs[0].set_xticklabels(xlabels, rotation=90)
axs[0].set_ylabel('Average reward')
axs[0].set_title('cooperative start')

axs[1].bar(results_h.keys(), results_h.values(), yerr=std_err_h, color='salmon')
axs[1].set_xticklabels(xlabels, rotation=90)
axs[1].set_title('hostile start')

plt.tight_layout()
plt.savefig('all_agents_earned_rewards.svg')
```

```python
# reviewer also wants to see: average number of wins & net gain (wins-losses).
# THIS IS: Average number of wins. 
# read those csv
df_wins_coop = pd.read_csv('wins_loss_coop.csv')
df_wins_coop = df_wins_coop.drop(['Unnamed: 0'], axis=1)

wins_c, net_wins_c = dict(), dict()
names = list(df_wins_coop.columns)
std_err_c = dict()

for player in names:
    wins = np.mean(ast.literal_eval(df_wins_coop[player][0]))
    loss = np.mean(ast.literal_eval(df_wins_coop[player][1]))
    
    wins_c[player] = wins
    net_wins_c[player] = wins-loss
    std_err_c[player] = np.std(ast.literal_eval(df_wins_coop[player][0])) 
```

```python
# same for hostile context
df_wins_host = pd.read_csv('wins_loss_hostile.csv')
df_wins_host = df_wins_host.drop(['Unnamed: 0'], axis=1)

wins_h, net_wins_h = dict(), dict()
names = list(df_wins_host.columns)
std_err_h = dict()

for player in names:
    wins = np.mean(ast.literal_eval(df_wins_host[player][0]))
    loss = np.mean(ast.literal_eval(df_wins_host[player][1]))
    
    wins_h[player] = wins
    net_wins_h[player] = wins-loss
    std_err_h[player] = np.std(ast.literal_eval(df_wins_host[player][0]))
```

```python
fig, axs = plt.subplots(1,2)
fig.suptitle(r'number of $win$ outcomes')

fig.set_figwidth(10)
fig.set_figheight(4)

axs[0].set_title('cooperative start')
axs[0].bar(list(wins_c.keys()), list(wins_c.values()), color='lightskyblue', yerr=list(std_err_c.values()))
axs[0].set_xticks(list(wins_c.keys()), xlabels_rev, rotation=90)
axs[0].set_ylabel('total nr. wins')
axs[0].set_ylim(-1,25)

axs[1].set_title('hostile start')
axs[1].bar(list(wins_h.keys()), list(wins_h.values()), yerr=list(std_err_h.values()), color='salmon')
axs[1].set_xticks(list(wins_c.keys()), xlabels_rev, rotation=90)
axs[1].set_ylim(-1,25)

plt.show()
```

```python
# next we have average net wins per agent type
xticks = list(range(0,5))
xticks_h = [x+0.3 for x in xticks]
plt.bar(list(range(0,5)), list(net_wins_c.values()), width=0.3, color='lightskyblue', label='coop. start')
plt.bar(xticks_h, list(net_wins_h.values()), width=0.3, color='salmon', label='host. start')
plt.title(r'Average net $wins$ (wins-loss)')
plt.xticks(xticks, list(net_wins_h.keys()), rotation=90)
plt.ylabel('Net wins')
plt.legend()
```
