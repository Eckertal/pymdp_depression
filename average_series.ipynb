# Average series plots per agent type

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import os, sys, glob
from vis import plot_average_series
```

```python
all_agents = ['Player1_healthy',
              'biased_A', 'biased_B', 'biased_C', 'biased_D',
              'Type1_depressed', 'Type2_depressed',
              'Type1_social_phobia', 'Type2_social_phobia']
              
os.chdir('C:\\Users\\annae\\Desktop\\pymdp_depression\\avg_series_results\\revision')
df_coop = pd.read_csv('posterior_coop_coopFirst.csv')
df_host = pd.read_csv('posterior_host_coopFirst.csv')
```

```python
def read_series_results(df, agent):
    '''pretty stupid way of reading the stuff...'''
    results_agent = df[agent]
    plot_res = []
    for item in results_agent:
        item = item.replace('[','')
        item = item.replace(']','')
        res = [float(idx) for idx in item.split(', ')]
        plot_res += [res]
    
    return plot_res
```

```python
for agent in all_agents: 
    # first make plots for coop first for all agents. 
    plot_res_coop = read_series_results(df_coop, agent)
    df_series_coop = pd.DataFrame(plot_res_coop)
    
    plot_res_host = read_series_results(df_host,agent)
    df_series_host = pd.DataFrame(plot_res_host)
    
    svg_name = str(f'{agent}_avg_pn.svg')
    plot_average_series(agent, plot_res_coop, plot_res_host, df_series_coop, df_series_host, svg_name, first_context = 'friendly')
    
```
