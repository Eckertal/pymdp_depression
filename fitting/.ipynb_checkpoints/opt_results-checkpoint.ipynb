```python
import os, sys, glob, pdb, pickle
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from utils import softmax_params, make_soft_B

root = os.getcwd()
data_path = os.path.join(root, 'data_trustgame_depression')
opt_res_path = os.path.join(root, 'opt_results')
fig_path = os.path.join(root, 'plots')

mother = os.path.dirname(root)
os.chdir(mother)
from gms import *
```

```python
os.chdir(data_path)
df = pd.read_csv('data_inclBDI_n25.csv')
subjects = list(df['Participant Private ID'].unique())
print(len(subjects))
```

```python
# read optim results
os.chdir(opt_res_path)
pkl_res = glob.glob("*.pkl")

# filter results to find all results that belong to one subject
sbj_results = {sbj: [] for sbj in subjects}
for sbj in subjects: 
    for res in pkl_res: 
        if f'_{sbj}_' in res: 
            sbj_results[sbj].append(res)
```

```python
best_pars = {sbj: [] for sbj in subjects}
min_cost_sbj = {}
n_fev = {}

for sbj in subjects:
    sbj_files = sbj_results[sbj]
    number_evals = []
    
    min_cost = 10000
    for f in sbj_files: 
        file = open(f, 'rb')
        run = pickle.load(file)
        cost = run['fun']
        number_evals.append(run['nfev'])
        if cost < min_cost:
            min_cost = cost
            best_pars[sbj] = run['x']
        n_fev[sbj] = number_evals
            
    min_cost_sbj[sbj] = min_cost
```

```python
mean_evals = {}
for sbj in subjects: 
    mean_sbj = np.mean(n_fev[sbj])
    mean_evals[sbj] = mean_sbj
```

```python
print('min evals', min(list(mean_evals.values())))
print('max evals', max(list(mean_evals.values())))
print('mean evals', np.mean(list(mean_evals.values())))
print('STD evals', np.std(list(mean_evals.values())))
```

```python
print('max cost',max(np.array(list(min_cost_sbj.values()))))
print('min cost',min(np.array(list(min_cost_sbj.values()))))
print('mean cost', np.mean(list(min_cost_sbj.values())))
print('SD cost', np.std(list(min_cost_sbj.values())))
```

```python
plt.figure(figsize=(8,5))
plt.title('Cost per participant')
string_keys = [str(key) for key in min_cost_sbj.keys()]
plt.bar(string_keys, min_cost_sbj.values(), color='coral')
#plt.ylim(0.0, 0.25)
_=plt.xticks(rotation=90)
_=plt.ylabel('cost')
plt.tight_layout()
os.chdir(fig_path)
plt.savefig('optimizer_costs.svg')
```

```python
soft_pars = {}
for item in best_pars.items():
    sbj_id = item[0]
    soft_sbj = softmax_params(*list(item[1]))
    soft_pars[sbj_id] = soft_sbj
    
df_p = pd.DataFrame.from_dict(soft_pars).T
```

```python
# catch BDI score and condition for each participant. 
bdi = dict()
conditions = dict()
for sbj in subjects:
    df_sbj = df[df['Participant Private ID'] == sbj]
    sbj_bdi = df_sbj['BDI'].mean()
    condition = df_sbj['Spreadsheet'].unique()[0]
    bdi[sbj] = sbj_bdi
    conditions[sbj] = condition

# determine BDI median split subjects
bdi_median = np.median(list(bdi.values()))
bdi_std = np.std(list(bdi.values()))

low_bdi = [sbj for sbj in subjects if bdi[sbj] < bdi_median]
high_bdi = [sbj for sbj in subjects if bdi[sbj] > bdi_median]

min_bdi = [sbj for sbj in subjects if bdi[sbj] < bdi_median - bdi_std]
max_bdi = [sbj for sbj in subjects if bdi[sbj] > bdi_median + bdi_std]
```

```python
low_c = [val for key, val in dict(df_p['C']).items() if key in low_bdi]
high_c = [val for key, val in dict(df_p['C']).items() if key in high_bdi]
all_c = [val for key, val in dict(df_p['C']).items()]

min_c = [val for key, val in dict(df_p['C']).items() if key in min_bdi]
max_c = [val for key, val in dict(df_p['C']).items() if key in max_bdi]
```

```python
# make a function to call the plots

def plot_c_violin(data, title=None, filename=None, A=False, D=False): 
    """
    data should be a list of list
    like low_c above. 
    """

    # Extract the 1st, 2nd, and 3rd elements from each sublist
    p_r0 = [x[0] for x in data]
    p_r1 = [x[1] for x in data]
    p_r2 = [x[2] for x in data]

    # Create a DataFrame to hold the data for plotting
    data = pd.DataFrame({
        'p_r0': p_r0,
        'p_r1': p_r1,
        'p_r2': p_r2
    })

    # violin plots
    plt.figure(figsize=(10, 4))
    if title==None: 
        plt.suptitle(r'$C$: optimized prior on outcomes')
    else:
        plt.suptitle(title)

    # win outcome
    plt.subplot(1, 3, 1)
    color='forestgreen'
    if A==True: 
        color='limegreen'
    ax = sns.violinplot(data=p_r0, color=color)
    for patch in ax.collections:
        patch.set_alpha(0.5)
    if A==True: 
        plt.title(r'$p(return)$|cooperative')
    if D==True: 
        plt.title(r'$p(cooperative)$')
    else: 
        plt.title('win')
    plt.ylim(0.0, 1.0)
    plt.ylabel(r'$p$')

    # loss outocme
    plt.subplot(1, 3, 2)
    color='coral'
    if A==True: 
        color='tomato'
    sns.violinplot(data=p_r1, color=color, alpha=0.2)
    if A==True: 
        plt.title(r'$p(return)$|hostile')
    if D==True: 
        plt.title(r'$p(hostile)$')
    else: 
        plt.title('loss')
    plt.ylim(0.0, 1.)

    # keep outcome
    plt.subplot(1, 3, 3)
    sns.violinplot(data=p_r2)
    
    if A==True: 
        plt.title(r'$p(return)$|random')
        
    if D==True: 
        plt.title(r'$p(random)$')
    else: 
        plt.title('keep budget')
    plt.ylim(0., 1.)
    
    if filename!=None:
        plt.savefig(filename)

    plt.tight_layout()
    plt.show()
```

# ϵ parameter

```python
epsilons = []
eps_dict = {}

for item in soft_pars.items(): 
    epsilons.append(item[1]['ep'])
for key, val in soft_pars.items(): 
    eps_dict[key] = val['ep']

fig, axs = plt.subplots(1,2)
fig.set_figheight(2)
fig.set_figwidth(4)
sns.violinplot(epsilons, ax=axs[0], color='skyblue')
axs[0].set_ylabel(r'$ϵ$')

axs[1].bar(range(0,len(epsilons)),epsilons, color='skyblue')
axs[1].set_xlabel('participant')

fig.suptitle(r'Weighting factor $\epsilon$')
plt.tight_layout()
plt.savefig('epsilon.svg')
```

```python
# difference in cost related to epsilon == 1/ 0? 
filtered_subjects_nonzero = {key for key, value in eps_dict.items() if value > 0.001}
filtered_costs = [min_cost_sbj[key] for key in filtered_subjects_nonzero if key in min_cost_sbj]
mean_cost = sum(filtered_costs) / len(filtered_costs) if filtered_costs else 0

print(f"The mean cost for subjects with epsilon > 0 is: {mean_cost}")

filtered_subjects_zero = {key for key, value in eps_dict.items() if np.isclose(value, 0)}
filtered_costs = [min_cost_sbj[key] for key in filtered_subjects_zero if key in min_cost_sbj]
mean_cost = sum(filtered_costs) / len(filtered_costs) if filtered_costs else 0

print(f"The mean cost for subjects with epsilon = 0 is: {mean_cost}")
```

```python
# separate for low/ high BDI
low_eps = [val for key, val in eps_dict.items() if key in low_bdi]
hi_eps  = [val for key, val in eps_dict.items() if key in high_bdi]

fig, axs = plt.subplots(1,2)
fig.set_figheight(2)
fig.set_figwidth(4)
axs[0].set_title('low BDI')
sns.violinplot(low_eps, ax=axs[0], color='powderblue', label='low BDI')
axs[1].set_title('high BDI')
sns.violinplot(hi_eps, ax=axs[1], color='cornflowerblue', label='high BDI')
plt.tight_layout()
plt.savefig('epsilon_low_high_BDI.svg')
plt.show()
```

# B parameter

```python
# B: transition matrix parameters. 
soft_B_sbj = dict()

for sbj in subjects: 
    params = list(dict(df_p['B'])[sbj])
    soft_B = make_soft_B(*params)
    soft_B_sbj[sbj] = soft_B  
```

```python
sbj_friendly, sbj_hostile, sbj_random = {},{},{}

for sbj in subjects:
    friendly = soft_B_sbj[sbj][:, :1, :].flatten()
    hostile  = soft_B_sbj[sbj][:, 1:2, :].flatten()
    random   = soft_B_sbj[sbj][:, 2:3, :].flatten()
    
    sbj_friendly[sbj] = friendly
    sbj_hostile[sbj]  = hostile
    sbj_random[sbj]   = random
```

```python
data = soft_B_sbj.copy()

# get average B results. 
array_shape = next(iter(data.values())).shape
mean_results = {}
std_results = {}
for array_index in range(array_shape[0]):
    for row_index in range(array_shape[1]):
        rows = [array[array_index, row_index] for array in data.values()]
        # Stack rows 
        stacked_rows = np.vstack(rows)
        # Compute mean
        mean_result = float(np.mean(stacked_rows, axis=0))
        std_result = float(np.std(stacked_rows, axis=0))
        # Store the mean result in the dictionary
        mean_results[(array_index, row_index)] = mean_result
        std_results[(array_index, row_index)] = std_result

B_labels = ['cc', 'hc', 'rc', 'ch', 'hh', 'rh', 'cr', 'hr', 'rr']
for key, value in zip(B_labels, mean_results.values()):
    print(f"{key}: Mean = {value}")
```
