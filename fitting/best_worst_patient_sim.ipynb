# Simulating the best and worst patient

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os, glob, sys, pdb, ast

fit_path = os.getcwd()
root = os.path.dirname(fit_path)

os.chdir(root)

from compute_avg_series import compute_series
from vis import plot_avg_series_revised
from library import get_player_agent
```

```python
df_coop = pd.read_csv('avg_series_pn.csv')
df_host = pd.read_csv('avg_series_np.csv')
```

```python
all_agents = ['healthiest', 'sickest'] #, 'sickest'


labels = {'healthiest': 'Patient with min. BDI score', 
         'sickest': 'Patient with max. BDI score'}

num = 1
for agent in all_agents: 
    title = labels[agent]

    # read results for coop-to-hostile game
    res_coop_pn = ast.literal_eval(df_coop[agent][0])
    res_host_pn = ast.literal_eval(df_coop[agent][1])
    # as dataframe
    df_series_coop_pn = pd.DataFrame(res_coop_pn)
    df_series_host_pn = pd.DataFrame(res_host_pn)

    # read results for host-to-coop game
    res_coop_np = ast.literal_eval(df_host[agent][0])
    res_host_np = ast.literal_eval(df_host[agent][1])
    # as dataframe
    df_series_coop_np = pd.DataFrame(res_coop_np)
    df_series_host_np = pd.DataFrame(res_host_np)

    svg_name = str(f'{num}_{agent}_avg_series.svg')
    num += 1

    plot_avg_series_revised(agent, res_coop_pn, res_host_pn, df_series_coop_pn, df_series_host_pn,
                                res_coop_np, res_host_np, df_series_coop_np, df_series_host_np,
                                title, svg_name)
```
