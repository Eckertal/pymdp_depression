# Plots

- time series data plots
- distribution of BDI values

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import seaborn as sns

fig_path = os.path.join(os.getcwd(), 'plots')
df = pd.read_csv('data_inclBDI_n17.csv')
```

```python
df_np = df[df['Spreadsheet']=='Negative-to-positive']
df_pn = df[df['Spreadsheet']=='Positive-to-negative']

sbj_np = df_np['Participant Private ID'].unique()
sbj_pn = df_pn['Participant Private ID'].unique()

for sbj in sbj_np:
    plt.figure(figsize=(10,2))
    df_sbj = df_np[df_np['Participant Private ID']==sbj]
    df_sbj.index = range(1,61)
    plt.step(range(1,61), df_sbj['Response'].values, color='black')
    plt.yticks([0,1], ['keep', 'invest'])
    plt.axvspan(0, 30, color='red', alpha=0.2, label='hostile')
    plt.axvspan(30, 60, color='blue', alpha=0.2, label='coop')
    plt.xlabel('trial')
    plt.title(sbj)
    partner_answer_1 = df_sbj[df_sbj['partnerAnswer'] == 1]
    plt.scatter(partner_answer_1.index + 1, partner_answer_1['partnerAnswer'], color='blue', marker='*', label='p. shared')
    partner_answer_0 = df_sbj[df_sbj['partnerAnswer'] == 0]
    plt.scatter(partner_answer_0.index + 1, partner_answer_0['partnerAnswer'], color='red', marker='x', label='p. abused')
    plt.legend(loc='upper right')
    os.chdir(fig_path)
    plt.savefig(f'{sbj}.svg')
    plt.show()
```

```python
bdi = pd.read_csv('bdi_n17.csv')
bdi = bdi.drop('Unnamed: 0', axis=1)
median_bdi = bdi['bdi_score'].median()
std_bdi = bdi['bdi_score'].std()
```

```python
filter_values = [0, 10, 20, 30, 40, 50, 60]

out = pd.cut(bdi['bdi_score'], bins = filter_values, include_lowest = True)
counts = out.value_counts(sort=False)


counts.plot.bar(rot=35, color='coral')
plt.title('Distribution of BDI scores')
plt.xlabel('BDI score')
plt.ylabel('Frequency')
plt.tight_layout()
os.chdir(fig_path)
plt.savefig('bdi_distribution.svg')
```
